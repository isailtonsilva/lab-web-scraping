```python
# Step 0 - Import libraries
# requests: download HTML pages
# BeautifulSoup: parse HTML
# pandas: store results in a DataFrame
# time/random: small delays so we don't spam the site

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
# Step 1 - Helper functions

def rating_to_number(rating_word):
    # Convert rating words (One..Five) into integers (1..5)
    rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
    return rating_map.get(rating_word, None)

# Step 1.2 (FIXED) - Convert price text into float safely
# Some pages show 'Â£51.77' instead of '£51.77' because of encoding issues
# We remove currency symbols and keep only digits and decimal point

def clean_price(price_text):
    # Convert to string and strip spaces
    price_text = str(price_text).strip()

    # Remove common currency artifacts: '£' and 'Â'
    price_text = price_text.replace("£", "").replace("Â", "").strip()

    # Keep only digits and decimal dot (extra safety)
    price_text = "".join(ch for ch in price_text if (ch.isdigit() or ch == "."))

    return float(price_text)
```

```python
# Step 2 - Scrape one page of books and return a list of dictionaries

def scrape_books_from_page(page_url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(page_url, headers=headers)

    if response.status_code != 200:
        return []

    soup = BeautifulSoup(response.text, "html.parser")
    book_cards = soup.select("article.product_pod")

    books = []

    for card in book_cards:
        title_tag = card.select_one("h3 a")
        title = title_tag["title"].strip() if title_tag else None

        relative_link = title_tag["href"] if title_tag and title_tag.has_attr("href") else None
        book_url = "https://books.toscrape.com/catalogue/" + relative_link.replace("../../../", "") if relative_link else None

        price_tag = card.select_one("p.price_color")
        price = clean_price(price_tag.get_text()) if price_tag else None

        rating_tag = card.select_one("p.star-rating")
        rating_word = None
        rating_num = None
        if rating_tag:
            classes = rating_tag.get("class", [])
            for c in classes:
                if c in ["One", "Two", "Three", "Four", "Five"]:
                    rating_word = c
                    break
            rating_num = rating_to_number(rating_word)

        availability_tag = card.select_one("p.instock.availability")
        availability = availability_tag.get_text(strip=True) if availability_tag else None

        books.append({
            "title": title,
            "price": price,
            "rating": rating_num,
            "availability": availability,
            "url": book_url
        })

    return books
```

```python
# Step 3 - Scrape multiple pages and filter by min rating and max price

def scrape_books(min_rating=3, max_price=20.0, pages=50):
    all_books = []

    for page_num in range(1, pages + 1):
        page_url = f"https://books.toscrape.com/catalogue/page-{page_num}.html"
        page_books = scrape_books_from_page(page_url)
        all_books.extend(page_books)

        time.sleep(random.uniform(0.5, 1.2))

    df = pd.DataFrame(all_books)
    df = df.dropna(subset=["price", "rating", "title"])

    df_filtered = df[(df["rating"] >= min_rating) & (df["price"] <= max_price)].copy()
    df_filtered = df_filtered.sort_values(by=["rating", "price"], ascending=[False, True]).reset_index(drop=True)

    return df_filtered
```

```python
# Step 4 - Run scraping with chosen filters

min_rating = 4
max_price = 30.0

books_df = scrape_books(min_rating=min_rating, max_price=max_price, pages=50)

books_df.head(10)
```

```python
# Step 5 - Save results to CSV (optional but good practice)

books_df.to_csv("books_filtered.csv", index=False)
```
